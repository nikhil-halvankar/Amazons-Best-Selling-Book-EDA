```python
# relation between user rating and reviews 

plt.figure(figsize=(15,6))
sns.scatterplot(x=data['User Rating'],y=data['Reviews'],data=data,hue=data['Genre'])
plt.show()
```

```python
# relation between user rating and price 

plt.figure(figsize=(15,6))
sns.scatterplot(x=data['User Rating'],y=data['Price'],data=data,hue=data['Genre'])
plt.legend(loc=2)
plt.show()
```

```python
x = data[data['User Rating']==4.9]
x.head(10)

y=x.groupby('Name').Reviews.mean().sort_values(ascending=False)
plt.figure(figsize=(15,15))
sns.barplot(y.values,y.index)
plt.show()
```

```python
# line graph to show change in reviews as time passes

plt.figure(figsize=(15,6))
sns.lineplot(x='Year',y='Reviews',data=data,hue='Genre')
plt.show()
```

```python
# line graph to show change in price as time passes

plt.figure(figsize=(15,6))
sns.lineplot(x='Year',y='Price',data=data,hue='Genre')
plt.show()
```

```python
# relation between each other

sns.pairplot(data,hue='Genre')
plt.show()
```

```python
# check for outliers

fig, axes=plt.subplots(1,3, figsize=(15,6))
data['User Rating'].plot(kind='box',ax=axes[0])
data['Reviews'].plot(kind='box',ax=axes[1])
data['Price'].plot(kind='box',ax=axes[2])
plt.show()
```
